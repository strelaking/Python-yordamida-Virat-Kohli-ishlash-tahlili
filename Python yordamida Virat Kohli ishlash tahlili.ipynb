Endi Python yordamida Virat Kohli unumdorligini tahlil qilish vazifasidan boshlaylik. Men ushbu vazifani kerakli Python kutubxonalari va ma'lumotlar to'plamini import qilish orqali boshlayman

Keling, oldinga siljishdan oldin ushbu ma'lumotlar to'plami null qiymatlarni o'z ichiga oladimi yoki yo'qligini ko'rib chiqamiz:

```python
# Total Runs Between 18-Aug-08 - 22-Jan-17
data["Runs"].sum()
```

Maʼlumotlar toʻplamida Virat Kohli 2008-yil 18-avgustdan 2017-yil 22-yanvargacha oʻynagan oʻyinlar mavjud. Keling, Virat Kohli tomonidan kiritilgan umumiy yugurishlarni koʻrib chiqaylik:

```python
# Average Runs Between 18-Aug-08 - 22-Jan-17
data["Runs"].mean()
```

Keling, xuddi shu davrdagi Virat Kohlining o'rtacha ko'rsatkichini ko'rib chiqaylik:


```python
matches = data.index
figure = px.line(data, x=matches, y="Runs", 
                 title='Runs Scored by Virat Kohli Between 18-Aug-08 - 22-Jan-17')
figure.show()
```

ODIlarda o'rtacha 35-37 ball yaxshi o'rtacha hisoblanadi. Shunday qilib, Virat Kohlning o'rtacha zarbasi yaxshi. Keling, Virat Kohlining 2008-yil 18-avgustdan 2017-yil 22-yanvargacha boʻlgan davrda oʻz karerasidagi gollari tendentsiyasini koʻrib chiqaylik:

```python
# Batting Positions
data["Pos"] = data["Pos"].map({3.0: "Batting At 3", 4.0: "Batting At 4", 2.0: "Batting At 2", 
                               1.0: "Batting At 1", 7.0:"Batting At 7", 5.0:"Batting At 5", 
                               6.0: "batting At 6"})

Pos = data["Pos"].value_counts()
label = Pos.index
counts = Pos.values
colors = ['gold','lightgreen', "pink", "blue", "skyblue", "cyan", "orange"]

fig = go.Figure(data=[go.Pie(labels=label, values=counts)])
fig.update_layout(title_text='Number of Matches At Different Batting Positions')
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=30,
                  marker=dict(colors=colors, line=dict(color='black', width=3)))
fig.show()
```

Virat Kohli o'ynagan ko'plab inninglarda u 100 dan ortiq yoki unga yaqin gol urdi. Bu izchillikning yaxshi belgisidir. Keling, Virat Kohli o'ynagan barcha zarbalarni ko'rib chiqaylik

```python
label = data["Pos"]
counts = data["Runs"]
colors = ['gold','lightgreen', "pink", "blue", "skyblue", "cyan", "orange"]

fig = go.Figure(data=[go.Pie(labels=label, values=counts)])
fig.update_layout(title_text='Runs By Virat Kohli At Different Batting Positions')
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=30,
                  marker=dict(colors=colors, line=dict(color='black', width=3)))
fig.show()
```

Virat Kohli o'ynagan barcha inninglarning 68% dan ortig'ida u uchinchi o'rinni egalladi. Keling, Virat Kohlining turli pozisiyalardagi umumiy yugurishlarini ko'rib chiqamiz:

```python
centuries = data.query("Runs >= 100")
figure = px.bar(centuries, x=centuries["Inns"], y = centuries["Runs"], 
                color = centuries["Runs"],
                title="Centuries By Virat Kohli in First Innings Vs. Second Innings")
figure.show()
```

```python
# Dismissals of Virat Kohli
dismissal = data["Dismissal"].value_counts()
label = dismissal.index
counts = dismissal.values
colors = ['gold','lightgreen', "pink", "blue", "skyblue", "cyan", "orange"]

fig = go.Figure(data=[go.Pie(labels=label, values=counts)])
fig.update_layout(title_text='Dismissals of Virat Kohli')
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=30,
                  marker=dict(colors=colors, line=dict(color='black', width=3)))
fig.show()
```

Virat Kohli tomonidan to'plangan umumiy yugurishlarning 72% dan ortig'i 3-o'rinda to'p surayotgan paytda. Demak, Virat Kohli uchun 3-pog'onada zarba berish juda mos keladi, deyishimiz mumkin.

Endi keling, Virat Kohli birinchi va ikkinchi inningda to'p o'ynatganda necha asrlik gol urganiga qaraylik:

```python
figure = px.bar(data, x=data["Opposition"], y = data["Runs"], color = data["Runs"],
            title="Most Runs Against Teams")
figure.show()
```

Shunday qilib, Virat Kohli tomonidan kiritilgan asrlarning aksariyati Avstraliyaga qarshi o'tdi. Endi Virat Kohlining zarba tezligini tahlil qilaylik. Virat Kohlining zarba tezligini tahlil qilish uchun men Virat Kohli o'ynagan barcha o'yinlarning yangi ma'lumotlar to'plamini yarataman, u erda uning zarbasi 120 dan ortiq edi:


```python
strike_rate = data.query("SR >= 120")
print(strike_rate)
```

Keling, Virat Kohli birinchi yoki ikkinchi inningda yuqori zarba stavkalari bilan o'ynashini ko'rib chiqaylik:

```python
figure = px.bar(strike_rate, x = strike_rate["Inns"], 
                y = strike_rate["SR"], 
                color = strike_rate["SR"],
            title="Virat Kohli's High Strike Rates in First Innings Vs. Second Innings")
figure.show()
```

Shunday qilib, yuqoridagi raqamga ko'ra, Virat Kohli ikkinchi inningga nisbatan birinchi inningda tajovuzkor o'ynashni yaxshi ko'radi. Keling, Virat Kohli tomonidan kiritilgan yugurishlar va uning har bir inningda o'ynagan to'rtliklari o'rtasidagi munosabatni ko'rib chiqamiz:

```python
figure = px.scatter(data_frame = data, x="Runs",
                    y="4s", size="SR", trendline="ols", 
                    title="Relationship Between Runs Scored and Fours")
figure.show()
```

Lineer munosabatlar mavjud. Bu shuni anglatadiki, Virat Kohli to'rtta o'ynashni yaxshi ko'radi. Inningda qanchalik ko'p yugursa, shuncha ko'p to'rtta o'ynaydi. Keling, oltitalar bilan qandaydir aloqasi bor-yo'qligini bilib olaylik:

```python
figure = px.scatter(data_frame = data, x="Runs",
                    y="6s", size="SR", trendline="ols", 
                    title= "Relationship Between Runs Scored and Sixes")
figure.show()
```

Bu erda kuchli chiziqli aloqa yo'q. Bu shuni anglatadiki, Virat Kohli oltitadan ko'proq to'rtta o'ynashni yaxshi ko'radi. Shunday qilib, siz Virat Kohli yoki dunyodagi boshqa kriketchilarning o'yinini tahlil qilishingiz mumkin.
